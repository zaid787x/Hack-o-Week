```python
import pandas as pd

data = []
n = int(input("Enter number of interactions: "))

for i in range(n):
    d = float(input(f"Enter distance between people {i+1} (in meters): "))
    data.append(d)

df_social = pd.DataFrame({
    "interaction": range(1, n+1),
    "distance": data
})

df_social
```

```python
def classify_interaction(d):
    if d < 1:
        return "RISKY"
    elif d < 2:
        return "CLOSE"
    else:
        return "SAFE"

df_social["status"] = df_social["distance"].apply(classify_interaction)

df_social
```

```python
risk_count = df_social["status"].value_counts()

print("Interaction Summary:")
print(risk_count)
```

```python
import matplotlib.pyplot as plt

df_social["status"].value_counts().plot(kind='bar')

plt.title("Social Interaction Risk Levels")
plt.xlabel("Category")
plt.ylabel("Count")

plt.show()
```
