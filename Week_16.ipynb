```python
import pandas as pd
import numpy as np

data = []
n = int(input("Enter number of sensor readings: "))

for i in range(n):
    val = input(f"Enter reading {i+1} (or press Enter for missing): ")

    if val == "":
        data.append(np.nan)   # missing value
    else:
        data.append(float(val))

df = pd.DataFrame({"distance": data})

print("Raw Data:")
df
```

```python
print("Missing values:")
print(df.isnull().sum())

# Fill missing with mean
df["distance"].fillna(df["distance"].mean(), inplace=True)

print("\nAfter filling missing values:")
df
```

```python
Q1 = df["distance"].quantile(0.25)
Q3 = df["distance"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df["distance"] >= lower) & (df["distance"] <= upper)]

print("After removing outliers:")
df
```

```python
df["distance"] = (df["distance"] - df["distance"].min()) / (df["distance"].max() - df["distance"].min())

print("Normalized Data:")
df
```
