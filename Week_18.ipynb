```python
import pandas as pd

data = []
n = int(input("How many readings do you want to enter? "))

for i in range(n):
    d = int(input(f"Enter distance for reading {i+1}: "))
    data.append(d)

df = pd.DataFrame({
    "time": range(1, n+1),
    "distance": data
})

df
```

```python
def classify(distance):
    if distance < 10:
        return "DANGER"
    elif distance < 20:
        return "WARNING"
    else:
        return "SAFE"

df["status"] = df["distance"].apply(classify)

df
```

```python
danger_cases = df[df["status"] == "DANGER"]

print("Danger Events:")
danger_cases
```

```python
import matplotlib.pyplot as plt

plt.plot(df["time"], df["distance"], marker='o')
plt.axhline(y=10, linestyle='--', label="Danger")
plt.axhline(y=20, linestyle='--', label="Warning")

plt.title("Warehouse Collision Detection")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.plot(df["time"], df["distance"], marker='o')

plt.title("Distance vs Time")
plt.xlabel("Time")
plt.ylabel("Distance")

plt.grid()

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.bar(df["time"], df["distance"])

plt.title("Distance Bar Chart")
plt.xlabel("Time")
plt.ylabel("Distance")

plt.show()
```

```python
status_count = df["status"].value_counts()

print(status_count)

status_count.plot(kind='bar')

plt.title("Safety Status Count")
plt.xlabel("Status")
plt.ylabel("Count")

plt.show()
```
